```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2015CountyTypologyCodes.csv')
df.tail(50)
```

```python
df.Economic_Type_Label.value_counts()
```

```python
# Creating the states dataset
statesData = pd.pivot_table(df, index='State', values='Economic Types Type_2015_Update non-overlapping', columns='Economic_Type_Label', aggfunc="count").reset_index()
statesData = statesData.rename_axis(None, axis=1)
statesData.head()
```

```python
# fill in NaN with 0 values
statesData.fillna(0, inplace=True)
statesData.head()
```

```python
# we calculate total number of counties per state
statesData['Total'] = statesData.sum(axis=1)
statesData.head()
```

```python
# we calculate the most common sector overall
statesData['max'] = statesData[['Farming', 'Federal/State Government', 'Maufacturing', 'Mining', 'Nonspecialized', 'Recreation']].idxmax(axis=1)
statesData.head(10)
```

```python
# we calculate the percentage (real representation) for each sector column
statesData.iloc[:,1:-2] = statesData.iloc[:,1:-2].div(statesData.Total, axis=0)
statesData.head(10)
```

```python
statesData =statesData.drop(['Total'], axis=1)
statesData.shape
```
